```python
def vectorizer(x):
    vector = {
 'M': 0,
 'E': 0,
 'S': 0,
 'L': 0,
 'V': 0,
 'P': 0,
 'G': 0,
 'F': 0,
 'N': 0,
 'K': 0,
 'T': 0,
 'H': 0,
 'Q': 0,
 'R': 0,
 'D': 0,
 'A': 0,
 'C': 0,
 'Y': 0,
 'I': 0,
 'W': 0}
    for i in x:
        vector[i] = vector[i]+1
    for i,k in zip(vector.keys(),vector.values()):
        vector[i] = k/len(x)
    return vector
```

```python
def diff(x,y):
    vector = {
 'M': 0,
 'E': 0,
 'S': 0,
 'L': 0,
 'V': 0,
 'P': 0,
 'G': 0,
 'F': 0,
 'N': 0,
 'K': 0,
 'T': 0,
 'H': 0,
 'Q': 0,
 'R': 0,
 'D': 0,
 'A': 0,
 'C': 0,
 'Y': 0,
 'I': 0,
 'W': 0}
    for key in vector.keys():
        if key in x.keys() and key in y.keys():
            if x[key]-y[key]>0:
                vector[key] = x[key]-y[key]
            else:
                vector[key] = 0
        elif key in x.keys() and key not in y.keys():
            vector[key] = x[key]
        elif key not in x.keys() and key in y.keys():
            vector[key] = -y[key]
    difference = 0
    for i in vector.values():
        difference += i
    vector["difference"] = "%" + str(difference*100)
    vector["similarity"] = "%" + str((1-difference)*100)
    return vector
```
